Importing, cleaning and merging Group 6's datasets

```python
import pandas as pd
```

```python
#import crime data
link="https://github.com/Sarahsid36/SPD/raw/main/SPDdata.csv"

crimedata=pd.read_csv(link)
```

```python
#import cultural spaces data
link="https://github.com/elstephens1216/PUBPOL542/blob/main/SeattleAggregatedCulturalSpaces.xlsx?raw=true"

culturalspaces=pd.read_excel(link)
```

```python
culturalspaces
```

```python
culturalspaces.info()
```

```python
#need to change zipcode to string (object) 
culturalspaces['zipcode'] = culturalspaces['zipcode'].apply(str)
```

```python
culturalspaces.dtypes
```

```python
#remove all zipcodes that don't start with 981
culturalspaces.zipcode.str.startswith('981')
```

```python
culturalspaces['inSeattle']=culturalspaces.zipcode.str.startswith('981')
```

```python
culturalspaces
```

```python
seattleZips=culturalspaces[culturalspaces.inSeattle==True]
```

```python
WhenYouFind="-"
ReplaceWith=""
```

```python
seattleZips.columns.str.replace(WhenYouFind,ReplaceWith,regex=True).str.lower()
```

```python
seattleZips = seattleZips.replace('-','', regex=True)
```

```python
seattleZips['zipcode']=seattleZips['zipcode'].str.strip()
```

```python
seattleZips
```

```python
df=seattleZips[seattleZips.inSeattle==True]
```

```python
df
```

```python
df.dtypes
```

```python
#turn zipcode back into integers
df['zipcode'].astype(str).astype(int)
```

```python
df.dtypes
```

```python
#merge crimedata and cultural spaces
crimedata.merge(df)
```

```python
merged1 = crimedata.merge(df)
```
